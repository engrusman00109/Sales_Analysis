## Data Visualization
## Exploratory Data Analysis:
</div> 

#### Gender Distribution :

```python
ax=sns.countplot(x="Gender",data=df)
for value in ax.containers:
    ax.bar_label(value,fontsize=10)
    
plt.title("Number of Male and Female in dataset",color="red",fontsize=15 )
plt.xlabel("Gender",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

#### Male Vs Female Sales:

```python
amount_wrt_gender=df.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
amount_wrt_gender
```

```python
ax=sns.barplot(x="Gender",y="Amount",data= amount_wrt_gender)
for value in ax.containers:
    ax.bar_label(value,fontsize=10)
    
plt.title("Sales with respect to Gender",color="red",fontsize=15 )
plt.xlabel("Gender",fontsize=15)
plt.ylabel("Amount",fontsize=15)
plt.show()
```

### Sales with respect to Age Group:

```python
sales_wrt_age=df.groupby(['Age Group'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_wrt_age
```

```python
sns.barplot(x="Age Group",y="Amount",data=sales_wrt_age)
plt.title("Sales with respect to Age group",color="red",fontsize=15 )
plt.show()
```

```python
sns.set(rc={"figure.figsize":(15,7)})
ax=sns.barplot(x="Age Group",y="Amount",data=sales_wrt_age)
for bar in ax.containers:
    ax.bar_label(bar,fontsize=10)
plt.xlabel("Age Group",fontsize=15)
plt.ylabel("Amount",fontsize=15)
plt.title("Sales with respect to Age group",color="red",fontsize=15 )
plt.show()
```

### Orders With Respect to Age Group and Gender :

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars,fontsize=20)
plt.title("Order with respect to Gender and Age Group",color="red",fontsize=15 )
plt.xlabel("Age Group",fontsize=20)
plt.ylabel("Number of Orders",fontsize=20)
plt.show()
```

### Orders with respect to State:

```python
sales_wrt_state=df.groupby(['State'],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
```

```python
sales_wrt_state
```

```python
sns.set(rc={"figure.figsize":(30,15)})
ax=sns.barplot(data=sales_wrt_state,x="State",y="Orders")
plt.title("Order with respect to State",color="red",fontsize=25 )
for bars in ax.containers:
    ax.bar_label(bars,fontsize=20)
plt.xlabel("States",fontsize=25)
plt.ylabel("Number of Orders",fontsize=25)
plt.show()
```

### Sales With respect to Marital_Status and Gender :

```python
sales_wrt_MaritalStatus=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_wrt_MaritalStatus
```

```python
sns.set(rc={"figure.figsize":(7,7)})
ax=sns.barplot(data=sales_wrt_MaritalStatus,x="Marital_Status",y="Amount",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Amount with respect to Gender",color="red",fontsize=15 )
plt.xlabel("Marital_Status",fontsize=15)
plt.ylabel("Amount",fontsize=15)
plt.show()
```

### Sales With Respect to Occupation:

```python
sales_wrt_Occupation=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_wrt_Occupation
```

```python
sns.set(rc={"figure.figsize":(30,10)})
ax= sns.barplot(data=sales_wrt_Occupation,x="Occupation",y="Amount")
for bars in ax.containers:
    ax.bar_label(bars,fontsize=17)
plt.xlabel("Occupation",fontsize=30)
plt.ylabel("Amount",fontsize=30)
plt.show()
```

### Number of Orders with respect to Occupation:

```python
sns.set(rc={"figure.figsize":(20,7)})
ax=sns.countplot(data=df,x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars,fontsize=15)
plt.title("Number of Order With respect to Occupation",color="red",fontsize=25 )
plt.xlabel("Occupation",fontsize=20)
plt.ylabel("Number of Orders",fontsize=20)
plt.show()
```
